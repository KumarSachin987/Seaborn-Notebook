# Point Plot

```python
tips = sns.load_dataset('tips')
tips.head(35)
```

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize = (15,8))
sns.pointplot(data=tips, x='time',y='total_bill',hue='smoker',dodge=True)
```

# Heat Map

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights = flights.pivot("month","year","passengers")
plt.figure(figsize = (15,8))
sns.heatmap(flights,annot=True,fmt='d',linewidths=.5)
```

# Stripplot

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize = (15,8))
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
plt.figure(figsize = (15,8))
sns.stripplot(data=tips,y='day',x='total_bill',jitter=True)
```

```python
plt.figure(figsize = (15,8))
sns.stripplot(data=tips,x='day',y='total_bill',linewidth=1,hue='sex')
```
